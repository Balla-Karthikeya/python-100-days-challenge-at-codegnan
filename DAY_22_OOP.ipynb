# When you assign a value using an object, Python creates or updates an instance attribute, NOT the class attribute.

# types of methods
# instance method - FIRST PARAMETER IS SELF ONLY UPDATE ON INSTANCE ATTRIBUTES
# class method    - @ class method - @ decatator
#                   def method_name(cls)  takes first parameter as class
# static method   - @ static method  
#                   def method_name()  does not take self or  class as first parameter
#                   We can not use instance attributes in a static method
#                   can work on method attributes
# abstract method

```python
class Student: 
    school="xyz school"   # class attribute   

    def __init__(self, name, age, cls): # __init__ it is a one type of initilize constructor 
        self.name=name     # instance attribute
        self.age=age       # instance attribute
        self.cls=cls       # instance attribute

    def details(self):  # we have to keep self                 
        return f"My name is {self.name} and {self.age} years old and belongs to {self.cls}"


    def update_school(cls, new_school):
        cls.school = new_school


s1=Student("ravi",14,9)  
s2=Student("srinu",15,10)   

s1.update_school("abc school")
print(s1.school)
print(s2.school)
```

```python
class Phonebook:
    ph_book = {}   
    def __init__(self, name: str = None, number: int = None):
        self.name = name
        self.number = number
    def add_contact(self):
        if self.name not in Phonebook.ph_book:
            Phonebook.ph_book[self.name] = self.number
            print(f"{self.name} contact saved successfully")
        else:
            print(f"{self.name} contact already exists")
    def update_contact(self):
        if self.name in Phonebook.ph_book:
            Phonebook.ph_book[self.name] = self.number
            print(f"{self.name} contact successfully updated")
        else:
            print(f"{self.name} contact not found")
    def delete_contact(self):
        if self.name in Phonebook.ph_book:
            del Phonebook.ph_book[self.name]
            print(f"{self.name} contact deleted successfully")
        else:
            print(f"{self.name} contact not found")
    def search_contact(self):
        if self.name in Phonebook.ph_book:
            print(f"{self.name}'s number is {Phonebook.ph_book[self.name]}")
        else:
            print(f"{self.name} contact not found")
    def display_all(self):
        if not Phonebook.ph_book:
            print("Phone book is empty")
        else:
            for name, number in Phonebook.ph_book.items():
                print(f"{name} : {number}")

c1 = Phonebook("Karthikeya", 9878676532)
c2 = Phonebook("ravi", 987654321)

c1.add_contact()
c2.add_contact()

c1.number = 111111111
c1.update_contact()

c3 = Phonebook("karthikya")
c3.search_contact()

c4 = Phonebook("ravi")
c4.delete_contact()

c5 = Phonebook()
c5.display_all()
```
