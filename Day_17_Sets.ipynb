```python
# set operations - union, intersection, difference, symmetric difference
# checking operations - subset - issubset, superset - superset, disjointset-isdisjoint
# add function 
```

```python
a={1,2,5,6}
b={4,5,6}
print(a.union(b)) # union creates a new set containing all elements from a and b
print(a.intersection(b)) # common elements
print(a.difference(b)) # a-b only return set a, which are not present in set b
# symmetric difference all elements from both sets which are not common
print(a.symmetric_difference(b))
print(a,b) # no change we are not updating
# union |
# intersection &
# difference -
```

```python
# subset and super set
a={1,2,3}
b={1,2,3,4,5,5}
print(a.issubset(b))
print(a.issuperset(b))
print(b.issuperset(a))

```

```python
# disjoint set
a={1,2}
b={2,1}
print(a.isdisjoint(b))
c={3,4}
print(a.isdisjoint(c))
```

```python
# book management using sets and a dictionary 
# set - stores book register number
# dictionary = it stores book details
# python - 101, author details
# java - 102, author details 
# book_id=set()

# operations 
# add book -  (id, name, author, quantity) check id in set if there ok or not there, then take id as key and remaining take it as values 
# delete book - (if id present in set, then only we can delete), then delete the id from set and book from dict
# total books we can use the length  function 
# all books 
# borrow a book 


# give starting default values  for id and book name
```

```python
book_id={101,102}
book_details={
   101: {"name": "python", "author": "karthikeya", "quantity":3},
   102: {'name': "java", "author": "karthik", "quantity":2}

}
def add_book(id,name,author,quantity):
    if id in book_id:
        print("book id is present")
    else:
        book_id.add(id)
        book_details[id]={
            "name": name,
            "author": author,
            "quantity": qty
        }
def del_book(id):
    if id in book_id:
        book_id.remove(id)
        book_details.pop(id)  
        print("book deleated succesfully")
    else:
        print("book not found")
def total_books():  
    print("total books: ", len(book_id))
def all_books():
    if not book_details:
        print("no books available")
    else:
        for id, details in book_details.items():
            print(f"book_id: {id}")
            print("name: ", details['name'])
            print("author: ", details["author"])
            print("quantity:", details["quantity"])
def borrow_book(id):
    if id in book_id:
        if book_details[id]["quantity"]>0:
            book_details[id]["quantity"]-=1
            print("book borrowed succesfully")
        else:
            print("book is out of stock")
    else:
        print("book id not found")
        
        

if __name__=="__main__":
    print("welcome to book management system project")
    while True:
        print("1. ADD BOOK")
        print("2. DELETE BOKK")
        print("3. TOTAL BOOKS")
        print("4. ALL BOOKS")
        print("5. BORROW BOOK")
        print("6. EXIT")
        choice=int(input("enter your choice: "))
        if choice == 1:
            id=int(input("enter the book id: "))
            name=input("enter the book name: ")
            author=input("enter the book author: ")
            qty=int(input("enter total no.of quantity: "))
            add_book(id,name,author,quantity)
        elif choice==2:
            id=int(input("enter book id: "))
            del_book(id)
        elif choice==3:
            total_books()
        elif choice==4:
            all_books()
        elif choice==5:
            id=int(input("enter book id: "))
            borrow_book(id)
        elif choice == 6:
            print("exit book management system")
            break

        else:
            print("invalid choice")
```
