####Types of arguments
1. positional arguments
2. keyword arguments
3. Default arguments
4. variable-length positional arguments
5. variable-length of keyword arguments

```python
# positional arguments
# subtraction of two numbers using functions
def sub(n1,n2):
    res=n1-n2
    return res
a=int(input())
b=int(input())
print(sub(a,b))  # n1- a, n2 -b
```

```python
def sub(n1,n2):
    res=n1-n2
    return res
a=int(input())
b=int(input())
print(sub(b, a))  # position of arguments changing the result is changing  to overcome this, we use keyword arguments
```

```python
def sub(n1,n2):
    res=n1-n2
    return res
a,b = 10,5
print(sub(a,b))
print(sub(b,a))
```

```python
# keyword arguments
def sub(n1,n2):
    res=n1-n2
    return res
a,b=10,5
print(sub(n1=a,n2=b))  # n1 = a, n2 = b
print(sub(n2=b,n1=a))
```

```python
numbers = input()   
def sum_of_number(numbers):
    total = 0
    for n in numbers:
        total = total + int(n)
    print(total)
sum_of_number(numbers)
```

```python
# read list of numbers and if number is even add 0.5 and if number is odd add 1
# lst = [1,2,3,4,5]
# res = [2, 2.5, 4, 4.5, 6]
lst = list(map(int, input().split()))

for i in range(len(lst)):
    if lst[i] % 2 == 0:      # even
        lst[i] = lst[i] + 0.5
    else:                   # odd
        lst[i] = lst[i] + 1

print(lst)
```
