```python
# write mode
a=open("sample.txt",'w')
string="ok"
a.write(string)
a.close()
print("done")
```

```python
# read mode
a=open("sample.txt", 'r')
print(a.read())
a.close()
```

```python
a=open("sample.txt", "a")
string=" okk"
a.write(string)
a.close()
```

```python
a=open("sample.txt", 'r')
print(a.read())
a.close()
```

```python
a=open("sample.txt" ,'r+')
string="python programming"
a.write(string)                             
a.write(' SQL query language ')
a.write(' python ')
a.seek(0)
print(a.read())
a.close()
```

```python
"""If we use write mode (w):
• If the file is new (not present) → it creates the file
• If the file already has content → it erases the old content"""
```

```python
"""
One-line memory rules (VERY IMPORTANT)
r+ → overwrite from start, leftovers may remain
w+ → wipe everything, then write
a+ → keep everything, add at end
In simple words:
r+  no erase, just replace
w+  full erase first
a+  only append, never touch old data
"""
```
