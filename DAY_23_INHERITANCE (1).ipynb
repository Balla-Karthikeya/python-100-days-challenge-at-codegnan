```python
# super example

class User:
    def __init__(self, name: str, email: str, passport, age):
        self.name = name
        self.passport = passport
        self.email = email
        self.age = age

    def details(self):
        return f"user name is {self.name}, email is {self.email}, passport is {self.passport} and {self.age} years old"

    def show(self):
        return f"user name is {self.name}, email is {self.email}"


class Admin(User):
    def __init__(self, name: str, email: str, passport: int, age: int, total_users: int):
        super().__init__(name, email, passport, age)
        self.total_users = total_users

    def users(self):
        return f"I am working as admin and total users are: {self.total_users}"


u1 = User(name="karthik", email="karthi123@gmail.com", passport=12345, age=30)
a1 = Admin(name="karthikeya", email="karthikeya123@gmail.com",
           passport=123456789, age=28, total_users=100)

print(a1.show())
```

```python
# types of inheritance
# single 
# multilevel
# multiple
# hybrid
# hierarchicaL 
```

```python
# Single Inheritance
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, roll_no):
        super().__init__(name)
        self.roll_no = roll_no

s1 = Student("ravi", 10)
print(s1.name, s1.roll_no)
```

```python
# multilevel
class Person:
    def __init__(self, name):
        self.name = name

class Employee(Person):
    def __init__(self, name, emp_id):
        super().__init__(name)
        self.emp_id = emp_id

class Manager(Employee):
    def __init__(self, name, emp_id, department):
        super().__init__(name, emp_id)
        self.department = department

m1 = Manager("rahul", 10, "IT")
print(m1.name, m1.emp_id, m1.department)
```

```python
# multiple
class Employee:
    def __init__(self, emp_id):
        self.emp_id = emp_id

class Student:
    def __init__(self, roll_no):
        self.roll_no = roll_no

class WorkingStudent(Employee, Student):
    def __init__(self, emp_id, roll_no):
        Employee.__init__(self, emp_id)
        Student.__init__(self, roll_no)

ws = WorkingStudent(1, 1)
print(ws.emp_id, ws.roll_no)
```

```python
# Hybrid 
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, roll_no):
        Person.__init__(self, name)
        self.roll_no = roll_no

class Employee(Person):
    def __init__(self, name, emp_id):
        Person.__init__(self, name)
        self.emp_id = emp_id

class Intern(Student, Employee):
    def __init__(self, name, roll_no, emp_id):
        Student.__init__(self, name, roll_no)
        Employee.__init__(self, name, emp_id)

i1 = Intern("Karthik", 10, 1)
print(i1.name, i1.roll_no, i1.emp_id)
```

```python
# Hierarchical 

class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, roll_no):
        Person.__init__(self, name)
        self.roll_no = roll_no

class Employee(Person):
    def __init__(self, name, emp_id):
        Person.__init__(self, name)
        self.emp_id = emp_id

s1 = Student("Karthik", 1)
e1 = Employee("Ravi", 2)

print(s1.name, s1.roll_no)
print(e1.name, e1.emp_id)
```
