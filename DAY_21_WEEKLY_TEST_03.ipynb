```python
# Write a Python program to find the sum of the digits of a given number.
n=123
s=0
while  n > 0:
    digit=n%10 #3
    n=n//10 #12
    s=s+digit
print(s)
```

```python
# Write a Python program to reverse a given number.
n=123
s=0
while n>0:
    digit=n%10
    n=n//10
    s=s*10+digit
print(s)
```

```python
# Write a Pythonprogram to check whether a given number is amstrong or not.
num = int(input())
sum = 0
temp = num
while temp > 0:
   digit = temp % 10
   sum += digit ** 3
   temp //= 10
if num == sum:
   print(num,"is an Armstrong number")
else:
   print(num,"is not an Armstrong number")
```

```python
# Write a Python program to find the number of digits in a given number.
n=123
count=0
while  n > 0:
    n=n//10
    count+=1
print(count)
```

```python
# Write a Python program to find the sum of the first n natural numbers.
n = int(input("Enter a number: "))
sum = 0
for i in range(1, n + 1):
    sum = sum + i
print(sum)
```

```python
# Write a Python program to find the sum of all elements in a list.
lst = [10, 20, 30]
total = 0
for i in lst:
    total = total + i
print("Sum of all elements in the list is:", total)
```

```python
# Program to count the frequency of each element in a list

lst = [1, 2, 3, 1, 2, 3, 3]
freq = {}
for i in lst:
    if i in freq:
        freq[i] = freq[i] + 1
    else:
        freq[i] = 1
for key in freq:
    print(key, "->", freq[key])
```
