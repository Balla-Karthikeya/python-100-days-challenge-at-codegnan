# Find topper details
# take the number of students as input
# for each student
   # take input as follows 
          # name 
          # marks - 6 subjects 
# who is topper - name, total marks, each subject marks
# average mark of total students
# each subject average marks 

```python
print("Welcome to find who is topper of the class")
n = int(input("Enter the number of students in the class: "))
names = []
marks = []
class_marks = []
percentage = []
print("Enter the all student marks")
for i in range(n):
    print(f"\nEnter the student {i+1} marks") 
    name = input("Enter the student name: ") 
    print("Enter the Telugu, English, Hindi, Maths, Science, Social marks, as per the order")
    Telugu, English, Hindi, Maths, Science, Social = map(int, input("Enter the student marks: ").split()[:6])  
    total_marks = Telugu + English + Hindi + Maths + Science + Social
    student_percentage = (total_marks / 600) * 100
    names.append(name)
    marks.append([Telugu, English, Hindi, Maths, Science, Social])
    class_marks.append(total_marks)
    percentage.append(student_percentage)
# Find topper
print("\nTopper in the class is")
top_marks = max(class_marks)
toppers_position = []
for i in range(n):
    if class_marks[i] == top_marks:
        toppers_position.append(i)
# Print topper details
for i in toppers_position:
    print("Name of the student is:", names[i])
    print("Marks of student is:", marks[i])
    print("Total Marks:", class_marks[i])
    print("Percentage:", percentage[i])
    print()
# Print all students details
print("All students details:")
for i in range(n):
    print(
        "Name:", names[i],
        "| Marks:", marks[i],
        "| Total Marks:", class_marks[i],
        "| Percentage:", percentage[i]
    )
```

# without using functions
# find the smallest value in the list   
# find the largest value in the list 
# len of the list 

```python
def find_largest(lst):
    if not lst:
        return "Error: List cannot be empty"
    largest = lst[0]                 
    for item in lst[1:]:             
        if item > largest:
            largest = item
    return largest
numbers = [10, 25, 3, 99, 45]
result = find_largest(numbers)
print(result)
```

```python
def find_smallest(lst):
    if not lst:
        return "Error: List cannot be empty"
    smallest = lst[0]                
    for item in lst[1:]:              
        if item < smallest:           
            smallest = item
    return smallest
numbers = [10, 25, 3, 99, 45]
result = find_smallest(numbers)
print(result)
```
