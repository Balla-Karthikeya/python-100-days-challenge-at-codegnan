```python
# palindrome of a string
a="hih"
b=a[::-1]
if a == b:
    print("palindrome")
else:
    print("not a palindrome")
```

```python
# Python program to check if the number is an Armstrong number or not
num = int(input())
sum = 0
temp = num
while temp > 0:
   digit = temp % 10
   sum += digit ** 3
   temp //= 10
if num == sum:
   print(num,"is an Armstrong number")
else:
   print(num,"is not an Armstrong number")
```
