```python
# username validation
     # minimum length - 8
     # at least one upper case and one lower case letter
     # first character is upper or lower
     # special character ('_','@')
     # it contains digits 
# password validation 
```

```python
s = input()

if len(s) >= 8:
    print("minimum length is satisfied")

    if s[0].isalpha():
        print("The first character is a letter")
    else:
        print("The first character must be a letter")

    if any(ch.isupper() for ch in s) and any(ch.islower() for ch in s):
        print("Contains both uppercase and lowercase letters")
    else:
        print("Must contain both uppercase and lowercase letters")

    if any(ch.isdigit() for ch in s):
        print("Contains digit")
    else:
        print("Must contain at least one digit")

    if any(ch in ['_', '@'] for ch in s):
        print("Contains special character (_ or @)")
    else:
        print("Must contain special character (_ or @)")

else:
    print("Username must be at least 8 characters long")     
```

```python
password = input()

if len(password) >= 8:
    print("Minimum length is satisfied")

    if password[0].isalpha():
        print("The first character is a letter")
    else:
        print("The first character must be a letter")

    if any(ch.isupper() for ch in password) and any(ch.islower() for ch in password):
        print("Contains both uppercase and lowercase letters")
    else:
        print("Must contain both uppercase and lowercase letters")

    if any(ch.isdigit() for ch in password):
        print("Contains digit")
    else:
        print("Must contain at least one digit")

    if any(ch in ['@', '#', '$', '_'] for ch in password):
        print("Contains special character (@, #, $, _)")
    else:
        print("Must contain at least one special character (@, #, $, _)")

else:
    print("Password must be at least 8 characters long")
```
