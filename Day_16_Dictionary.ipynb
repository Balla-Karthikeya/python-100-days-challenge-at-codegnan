```python
# DICTIONARY
# It is an ordered collection of key-value pairs. It is mutable and represented with {}
# properties:
# The key should be unique 
# values allow only duplicates
# list can't be used as a key
# All set elements can be used as dictionary keys
# dictionary is mutable 
# dictionary not index-based and not sliceable
# dictionary values can be accessed by key
# key and value separated with colon(:)
# {'name': "karthikeya", "age":20}
# items are separated by (,)
```

```python
# dictionary functions
# accesing values
# get(key, default_value): it returns the value if it is present in the dictionary, if not default value
details={'name': "karthikeya", "age":21, "batch":47, "course": 'python'}
print(details.get('name'))
print(details.get('percentage')) # return none
print(details.get('percentage',90))
```

```python
# updating or adding items to a dictionary 
# update(new_dictionary): it updates with new_dictionary items
details={'name': "karthikeya", "age":21, "batch":47, "course": 'python', 'percentage':85}
marks={"python":90, "soft skills": 85, 'aptitude ': 78,'percentage': 90}
details.update(marks)
print(details)
print(marks)
```

```python
# dictionary does not allow duplicates
details={'name': "karthikeya", "age":21, 'percentage':87, "batch":47, "course": 'python', 'percentage':85, 'percentage':90}
print(details)
```

```python
# get dictionary keys and values
# keys() - it returns a list of all dictionary keys
# values() - it returns a list of all dictionary values
# items()  - it returns all dicitonary items as list of tuple of key-value pairs
details={'name': "karthikeya", "age":21, 'percentage':87, "batch":47, "course": 'python', 'percentage':85, 'percentage':90}
print(details.keys())
print(details.values())
print(details.items())
print(type(details.keys()))
print(type(details.values()))
print(type(details.items()))
print(list(details.keys()))
```

```python
# deleting items from a dictionary 
# pop(key, default_value) - it returns bvalue and removes the specific key item from the dictionary
# popitem() - it returns and removes the last item from the dictionary
details={'name': "karthikeya", "age":21, "batch":47, "course": 'python'}
poped_value=details.pop('age')
print(poped_value)
print(details)
poped_value=details.pop('age',None)
print(poped_value)
print(details)
item=details.popitem()
print(item)
print(details)
```

```python
# contacts book 
# new contact 
# update mobile num
# delete contact
# get contact number
# print all contacts(name,no)
```

```python
ph_book = {}

def add_contact(name, num):
    if name not in ph_book:
        ph_book[name] = num
        print("Details saved")
    else:
        print("Contact already exists")

def update_contact(name, num):
    if name in ph_book:
        ph_book[name] = num
        print("Contact updated")
    else:
        print("Contact not found")

def del_contact(name):
    if name in ph_book:
        ph_book.pop(name)
        print("Contact deleted successfully")
    else:
        print("Contact not found")

def get_contact(name):
    if name in ph_book:
        print("Number:", ph_book[name])
    else:
        print("Contact not found")

def all_contacts():
    if not ph_book:
        print("Phone book is empty")
    else:
        for name, num in ph_book.items():
            print(f"{name} : {num}")

if __name__ == "__main__":
    print("Welcome to Phone book project")

    while True:
        print("\n1. ADD CONTACT")
        print("2. UPDATE CONTACT")
        print("3. DELETE CONTACT")
        print("4. FIND CONTACT")
        print("5. ALL CONTACTS")
        print("6. EXIT")

        choice = int(input("Enter the number from 1 to 6: "))

        if choice == 1:
            name = input("Enter name: ")
            num = int(input("Enter number: "))
            add_contact(name, num)

        elif choice == 2:
            name = input("Enter name: ")
            num = int(input("Enter number: "))
            update_contact(name, num)

        elif choice == 3:
            name = input("Enter name: ")
            del_contact(name)

        elif choice == 4:
            name = input("Enter name: ")
            get_contact(name)

        elif choice == 5:
            all_contacts()

        elif choice == 6:
            print("Exiting Phone Book. Bye ")
            break

        else:
            print("Invalid choice. Please enter 1 to 6.")
```
