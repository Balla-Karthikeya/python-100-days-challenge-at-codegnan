```python
# variable length of positional arguments
def add(a, *args):
    print(a)        #inside func tuple
    print(args)
print(add(1,2,3,4,5,6))
```

```python
# variable length of positional arguments
def add(a, *args):
    print(a)        #inside func tuple
    print(args)
    s=0
    for num in args:
        s += num
    return s
add(1,2,3,4,5,6)
```

```python
# convert list of numbers into  even list and odd list
lst=list(map(int,input().split()))
even_lst=[]
odd_lst=[]
for numbers in lst:
    if numbers %2:
        odd_lst.append(numbers)
    else:
        even_lst.append(numbers)
print(even_lst)
print(odd_lst)
```

```python
# mark analyser 
# in the list contains student marks, in that, find the pass percentage and fail percentage 
# pass marks are 35+
# fail marks are below 35
lst=list(map(int,input().split()))
pass_lst=[]
fail_lst=[]
for numbers in lst:
    if numbers > 35:
        pass_lst.append(numbers)
    else:
        fail_lst.append(numbers)
print(pass_lst)
print(fail_lst)
```
